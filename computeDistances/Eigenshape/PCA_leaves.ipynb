```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
%matplotlib inline
```

```python
leaves = np.genfromtxt('/home/marslast/Projects/AllPots/Data/Leaves/Leaves_Final_Coords_Scaled.csv', delimiter=',')
nleaves = len(leaves)
npoints = np.shape(leaves)[1]//2
print(nleaves,npoints)
#pl.plot(leaves[339,::2],leaves[339,1::2])
```

```python
# Perform PCA
C = np.cov(leaves.T)

# Get the eigenvalues and eigenvectors
evals,evecs = np.linalg.eig(C)

# Now need to sort them into descending order
indices = np.argsort(evals)
indices = indices[::-1]
evecs = evecs[:,indices]
evals = evals[indices]
evecs = np.real(evecs)
evals = np.real(evals)
m = np.mean(leaves,axis=0).reshape((npoints,2))
#evecs = evecs.reshape((npoints,2,npoints*2))
```

```python
# Plot a principal component
evecs2 = evecs.reshape((npoints,2,npoints*2))
sd = 1
pc = 2
pl.plot(m[:,0],m[:,1],'b')
pl.plot(m[:,0] - np.sqrt(evals[pc])*sd*evecs2[:,0,pc],m[:,1] - np.sqrt(evals[pc])*sd*evecs2[:,1,pc],'r') 
pl.plot(m[:,0] + np.sqrt(evals[pc])*sd*evecs2[:,0,pc],m[:,1] + np.sqrt(evals[pc])*sd*evecs2[:,1,pc],'r') 
```

```python
# Scree plot of the cumulative sum of variances
cs = np.cumsum(evals)
cs /= cs[-1]
pl.plot(cs)
varToSave = 0.85
print("Number of dims to use:",np.where(cs>varToSave*cs[-1])[0][0])
```

```python
# Plot of the first three principal components
for pc in range(3):
    for sd in range(1,4):
        pl.subplot(3,3,pc*3+sd);
        pl.plot(m[:,0],m[:,1],'b')
        pl.plot(m[:,0] - np.sqrt(evals[pc])*sd*evecs2[:,0,pc],m[:,1] - np.sqrt(evals[pc])*sd*evecs2[:,1,pc],'r') 
        pl.plot(m[:,0] + np.sqrt(evals[pc])*sd*evecs2[:,0,pc],m[:,1] + np.sqrt(evals[pc])*sd*evecs2[:,1,pc],'g')         
        pl.axis('off')
```

```python
# Plot according to PCA locations in 2D

# This sets up the colours for PCA plotting          
ind = np.loadtxt('/home/marslast/Projects/AllPots/Code/leaf_list.txt')
cl = np.unique(ind[:,0]).astype(int)

import matplotlib.colors as plc
cmap = pl.cm.get_cmap('inferno')
cnorm = plc.Normalize(vmin=0.0,vmax=len(cl))
colours = {}

count = 0
for i in cl:
    colours[i] = cmap((cnorm(count)))
    count += 1

def meanshape(leaves):
    means = np.zeros((15,npoints*2))
    for i in range(len(cl)):
        which = np.where(ind[:,0]==i+1)
        means[i,:] = np.mean(leaves[which,:],axis=1)
    return means

def plotpca(pca,leaves,x,y,scaling=None,shift=None):
    means = meanshape(leaves)
    pl.figure(figsize=(20,20));
    if scaling is None:
        scaling=1
    print(scaling)
    if shift is None:
        shift = np.array([0,0])
    for i, (pr, x, y) in enumerate(zip(pca, x,y)):
         pl.plot(pr[0],pr[1],'.',color=colours[int(ind[i,0])],markersize=10);
    for i in range(len(cl)):
        which = np.where(ind[:,0]==i+1)
        mean = np.squeeze(np.median(pca[which,:],axis=1))
        pl.plot(mean[0]+ means[i,::2]/scaling,mean[1] + means[i,1::2]/scaling,color=colours[i+1],linewidth=3);
                        
        #pl.text(-pr[0]+shift[0],pr[1]+shift[1],i, bbox={'facecolor':'white', 'edgecolor': 'white', 'alpha':.5});
```

```python
ndim = 2
m = np.mean(leaves,axis=0)
b = np.zeros((nleaves,ndim))
for i in range(nleaves):
    P = evecs[:,:ndim]
    b[i,:] = np.dot(P.T,(leaves[i,:]-m))

leaves2 = leaves.reshape(nleaves,npoints,2)

plotpca(b,leaves,leaves2[:,:,0],leaves2[:,:,1],scaling=1.5)
pl.axis('equal')
pl.axis('off')
pl.title('PCA');
pl.savefig('Leaves_PCA.pdf',dpi=600)
```
